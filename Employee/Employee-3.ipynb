```python
results = []

group0 = data[data['attrition_bin'] == 0]
group1 = data[data['attrition_bin'] == 1]

for col in numerical_cols:
    t_stat, p_val = ttest_ind(group0[col], group1[col], equal_var=False)
    results.append({
        'Признак': col,
        'Среднее (Stayed)': group0[col].mean(),
        'Среднее (Left)': group1[col].mean(),
        'p-value': p_val
    })
    
ttest_df = pd.DataFrame(results).sort_values('p-value').sort_index()
ttest_df['importance'] = ttest_df['p-value'].apply(lambda p: 'Статистически значим' if p < 0.05 else 'Нет статистической значимости')
ttest_df = ttest_df.loc[:6, ['Признак', 'importance']]

def cohen_d(x, y):
    
    pooled_std = np.sqrt(((x.std() ** 2) + (y.std() ** 2)) / 2)
    return (x.mean() - y.mean()) / pooled_std

ttest_df['cohen_d'] = [
    cohen_d(group0[col], group1[col])
    for col in ttest_df['Признак']
]

ttest_df['effect'] = ttest_df['cohen_d'].apply(lambda x: 'большой эффект' if abs(x) > 0.8 else
              'средний эффект' if abs(x) > 0.5 else
              'маленький эффект' if abs(x) > 0.2 else
              'пренебрежимо малый эффект'
)
display(ttest_df)
```

```python
def cohen_d_ci(x, y, alpha=0.05):
    d = cohen_d(x, y)
    n0, n1 = len(x), len(y)
    
    se = np.sqrt((n0 + n1) / (n0 * n1) + (d ** 2) / (2 * (n0 + n1)))
    z = scipy.stats.norm.ppf(1 - alpha / 2)
    ci_low = d - z * se
    ci_high = d + z * se
    return d, ci_low, ci_high
```

```python
ci_results = []

for col in ttest_df['Признак']:
    d, low, high = cohen_d_ci(group0[col], group1[col])
    ci_results.append({'Признак': col, 'cohen_d': d, 'CI_lower': low, 'CI_upper': high})

ci_df = pd.DataFrame(ci_results)
display(ci_df)
```

```python
ci_df_sorted = ci_df.sort_values(by='cohen_d', key=abs, ascending=False)

plt.figure(figsize=(10, 6))
plt.errorbar(
    x=ci_df_sorted['cohen_d'],
    y=ci_df_sorted['Признак'],
    xerr=[ci_df_sorted['cohen_d'] - ci_df_sorted['CI_lower'], 
          ci_df_sorted['CI_upper'] - ci_df_sorted['cohen_d']],
    fmt='o', color='black', ecolor='skyblue', capsize=5, label='Доверительный интервал'
)

plt.axvline(x=0, color='red', linestyle='--', label='Нет эффекта')

plt.xlabel("Эффект Коэна (Cohen's d)")
plt.ylabel("Признак")
plt.title("Доверительные интервалы эффекта Коэна по признакам")
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

**Вовод по исследовательскому анализу (EDA)**

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r) - 1)))

numeric_cols = data.select_dtypes(include='number').columns.drop(['attrition'], errors='ignore')
num_corr = data[numeric_cols].corrwith(data['attrition']).abs()
num_corr = pd.DataFrame({'feature': num_corr.index, 'correlation': num_corr.values})

cat_cols = data.select_dtypes(include='object').columns
cat_corr = {col: cramers_v(data[col], data['attrition']) for col in cat_cols}
cat_corr = pd.DataFrame({'feature': list(cat_corr.keys()), 'correlation': list(cat_corr.values())})

all_corr = pd.concat([num_corr, cat_corr], ignore_index=True)
all_corr = all_corr.sort_values(by='correlation', ascending=False)

plt.figure(figsize=(16, 8))
sns.barplot(
    data=all_corr,
    x='feature',
    y='correlation',
    palette=sns.color_palette('rocket', n_colors=len(all_corr)) 
)
plt.axhline(y=0.05, color='red', linestyle='-')
plt.xticks(rotation=45, ha='right')
plt.title('Сила связи признаков с целевой переменной (attrition)', fontsize=14)
plt.ylabel('Значение корреляции / силы связи')
plt.tight_layout()
plt.show()
```

```python
target = 'attrition'

RANDOM_STATE = 42
TEST_SIZE = 0.2

X = data.drop(columns=[target], axis = 1) 
y = data[target]

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=TEST_SIZE, 
                                                    stratify=y, 
                                                    random_state=RANDOM_STATE)
```

```python
cat_col = X.select_dtypes(include='object').columns.tolist()

ord_col = ['work-life_balance', 'job_satisfaction', 'performance_rating', 'company_reputation']
ohe_col = [col for col in cat_col if col not in ord_col]
num_col = X.select_dtypes(exclude='object').columns.tolist()

print(ord_col)
print(ohe_col)
print(num_col)
```

```python
ohe_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

ord_pipe = Pipeline([
    ('imputer1', SimpleImputer(strategy='most_frequent')),
    ('ord', OrdinalEncoder(categories=[
        ['Poor', 'Fair', 'Good', 'Excellent'],   
        ['Low', 'Medium', 'High', 'Very High'], 
        ['Low', 'Below Average', 'Average', 'High'], 
        ['Poor', 'Fair', 'Good', 'Excellent']   
    ], handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('imputer2', SimpleImputer(strategy='most_frequent'))
])

num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

data_preprocessor = ColumnTransformer([
    ('ohe', ohe_pipe, ohe_col),
    ('ord', ord_pipe, ord_col),
    ('num', num_pipe, num_col)
], remainder='drop')

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('feature_selection', SelectKBest(score_func=f_classif, k=10)),
    ('models', LogisticRegression())
])

param_grid = [
    {
        'models': [CatBoostClassifier(verbose=0)],
        'models__depth': [6],  
        'models__learning_rate': [0.1],
        'models__iterations': [100],
        'feature_selection__k': ['all']  
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear')],
        'models__C': [1, 5],  
        'preprocessor__num__scaler': [StandardScaler()],
        'feature_selection__k': ['all']
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='linear', probability=True)],
        'models__C': [1],  
        'preprocessor__num__scaler': [StandardScaler()],
        'feature_selection__k': ['all']
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': [5], 
        'models__weights': ['uniform'],
        'preprocessor__num__scaler': [StandardScaler()],
        'feature_selection__k': ['all']
    }
]
```

```python
grid = GridSearchCV(
    pipe_final,
    param_grid,
    cv=3,
    scoring='recall',
    n_jobs=-1,
    verbose=2
)
grid.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(grid.cv_results_)

best_per_model = (
    cv_results
    .assign(model_name=cv_results['param_models'].apply(lambda x: x.__class__.__name__))
    .sort_values('mean_test_score', ascending=False)
    .drop_duplicates(subset='model_name', keep='first')
    [['model_name', 'mean_test_score']]
    .rename(columns={
        'model_name': 'Model',
        'mean_test_score': 'Recall (CV)'
    })
    .reset_index(drop=True)
)

best_per_model = best_per_model.sort_values('Recall (CV)', ascending=False).reset_index(drop=True)

display(best_per_model)

best_model_name = best_per_model.iloc[0]['Model']
display(Markdown(f"**Лучшая модель по метрике Recall: `{best_model_name}`**"))
```

```python
pipe_final_cat = Pipeline([
    ('preprocessor', data_preprocessor),
    ('feature_selection', SelectKBest(score_func=f_classif, k=10)),
    ('model', CatBoostClassifier(verbose=0, random_state=RANDOM_STATE))
])
param_grid_cat = {
    'model__depth': [4, 6, 8],
    'model__learning_rate': [0.03, 0.05, 0.1],
    'model__iterations': [100, 200],
    'model__l2_leaf_reg': [1, 3, 5],
    'feature_selection__k': [5, 10, 'all']
}
```

```python
grid_cat = GridSearchCV(
    pipe_final_cat,
    param_grid_cat,
    scoring='recall',
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_cat.fit(X_train, y_train)
```

```python
print(grid_cat.best_params_)
print(grid_cat.best_score_)
```

```python
y_proba = grid_cat.predict_proba(X_test)[:, 1]
thresholds = np.arange(0.0, 1.01, 0.01)
results = []

for thresh in thresholds:
    y_pred = (y_proba >= thresh).astype(int)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    results.append((thresh, recall, precision))

df_metrics = pd.DataFrame(results, columns=['threshold', 'recall', 'precision'])

df_filtered = df_metrics[df_metrics['recall'] >= 0.8]

best_row = df_filtered.loc[df_filtered['precision'].idxmax()]
print("Лучший порог:", best_row)

plt.figure(figsize=(10, 5))

plt.plot(df_metrics['threshold'], df_metrics['recall'], label='Recall', color='blue')
plt.plot(df_metrics['threshold'], df_metrics['precision'], label='Precision', color='green')

plt.axvline(x=best_row['threshold'], color='red', linestyle='--', label=f'Лучший порог: {best_row["threshold"]:.2f}')

plt.xlabel('Порог')
plt.ylabel('Значение метрики')
plt.title('Recall и Precision в зависимости от порога')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.inspection import permutation_importance
best_pipeline = grid_cat.best_estimator_

X_test_transformed = best_pipeline.named_steps['feature_selection'].transform(
    best_pipeline.named_steps['preprocessor'].transform(X_test)
)

all_features = best_pipeline.named_steps['preprocessor'].get_feature_names_out()
mask = best_pipeline.named_steps['feature_selection'].get_support()
selected_features = all_features[mask]

result = permutation_importance(
    best_pipeline.named_steps['model'],
    X_test_transformed,
    y_test,
    n_repeats=10,
    scoring='recall',
    n_jobs=-1,
    random_state=42
)

importance_df = pd.DataFrame({
    'feature': selected_features,
    'importance': np.abs(result.importances_mean)
}).sort_values(by='importance', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(data=importance_df, x='importance', y='feature', hue=None, palette='viridis')
plt.title('Важность признаков')
plt.xlabel('Среднее снижение метрики recall')
plt.ylabel('Признак')
plt.tight_layout()
plt.show()
```
